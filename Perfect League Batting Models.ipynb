```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn import linear_model
import pickle
np.set_printoptions(precision=3, suppress=True)
```

```python
vsr = pd.read_csv('D:/perfect_team/data/vsrhp_lg_batting.csv', index_col='card_id')
vsr.columns
```

```python
r_vars = vsr[['batting_ratings_vsr_babip',
       'batting_ratings_vsr_gap', 'batting_ratings_vsr_power',
       'batting_ratings_vsr_eye', 'batting_ratings_vsr_strikeouts',
       'running_ratings_speed', 'bats_1', 'bats_2', 'bats_3']]

poly = PolynomialFeatures(2)
scaler = StandardScaler()

r_vars = scaler.fit_transform(poly.fit_transform(r_vars))

r_vars
```

```python
target = 'bb'

lr = linear_model.Ridge()

lr.fit(r_vars, vsr[target])

a = plt.axes(aspect='equal')
plt.scatter(vsr[target], lr.predict(r_vars))
plt.xlabel('True Values [HR_Rate]')
plt.ylabel('Predictions [HR_Rate]')
lims = [0.00, .20]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)

```

```python
diff = pd.DataFrame(vsr[target] * 600)
diff['pred'] = lr.predict(r_vars) * 600
diff['diff'] = diff[target] - diff['pred']
diff['diff'].hist()
```

```python
pickle.dump(lr, open('d:/perfect_team/python/batting/models/league/' + target + '_lhp', 'wb'))
```
